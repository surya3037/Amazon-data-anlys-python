```python
# to change the datatype

df['ship-postal-code']=df['ship-postal-code'].astype('int')
```

```python
# also to change the datatype of date

df['Date']=pd.to_datetime (df['Date'])
```

```python
#size

ax=sns.countplot(x='Size',data=df)
```

```python
# to presnatate the labels of bars

ax=sns.countplot(x='Size',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# group by

df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
```

```python
S_Quantity=df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)

ax=sns.barplot(x='Size',y='Quantity' , data=S_Quantity)
```

```python
sns.countplot(data=df, x='Courier Status',hue='Status')
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df, x='Courier Status', hue='Status')
plt.show()
```

```python
#histogram

df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=45)
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
plt.show()
```

```python
# here check the B2B by using pie chart

B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

```python
# scatterplot 

x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data, y_data, color='green')
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('scatter plot')
plt.show()
```

```python
#plot count of states cities

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('Ship-state')
plt.ylabel('count')
plt.title('distrubution of states')
plt.xticks(rotation=90)
plt.show()
```

```python

top_10_state = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12, 6))

sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')

plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```
